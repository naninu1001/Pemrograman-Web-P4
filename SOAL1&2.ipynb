### **1. Praktikum 4.3 – Linked Style Sheet**

Pada praktikum ini, semula pada baris ke-5 terdapat kode:

```html
<link rel="stylesheet" type="text/css" href="4.3 Linked style sheet.css">
```

Kemudian saya ubah sesuai instruksi soal menjadi:

```html
<link rel="stylesheet" type="text/css" href="#">
```

Setelah dijalankan di browser, halaman tetap menampilkan teks **TANAMAN HIAS ANYELIR**, **ANYELIR**, dan paragraf deskripsi, tetapi **tanpa warna, tanpa background, dan tanpa format** seperti sebelumnya.

### **Hasil Pengamatan dan Penjelasan**

Ketika atribut `href="#"` digunakan, browser **tidak menemukan file CSS** yang dimaksud karena tanda `#` bukan alamat file yang valid. Akibatnya, **file stylesheet tidak berhasil dimuat**, dan semua gaya (warna, tata letak, ukuran font, serta background gambar) **tidak diterapkan** ke elemen HTML.

Jadi halaman yang tampil hanyalah **struktur HTML polos** tanpa desain dari CSS.

**Kesimpulan:**  
Perubahan pada atribut `href` menjadi `#` menyebabkan file CSS tidak terbaca, sehingga tampilan halaman kembali ke bentuk dasar HTML tanpa style.

### **2. Praktikum 4.7 – Dropdown Menu dan Halaman About Me**

Pada bagian ini saya membuat menu dropdown seperti latihan 4.7, dengan menambahkan daftar materi praktikum 4.1 sampai 4.7 di dalam submenu **Praktikum 4**, serta menambahkan menu **About Me** yang berisi profil dan foto pribadi.

Berikut hasil pengerjaan saya:

#### **a. File: 4.7 Dropdown Menu.html**

```html
<!DOCTYPE html>
<html>
  <head>
    <title>Latihan 7</title>
    <link rel="stylesheet" type="text/css" href="4.7 Dropdown Menu.css">
  </head>
  <body>
    <div id="wrapper">
      <h3>Materi Praktikum Pemrograman Web</h3>
      <div id="menu">
        <ul>
          <li><a href="#">Praktikum 1</a></li>
          <li><a href="#">Praktikum 2</a></li>
          <li><a href="#">Praktikum 3</a></li>
          <li><a href="#">Praktikum 4</a>
            <ul>
              <li><a href="4.1 Embedded style sheet.html">4.1 Embedded Style Sheet</a></li>
              <li><a href="4.2 Inline style sheet.html">4.2 Inline Style Sheet</a></li>
              <li><a href="4.3 Linked style sheet.html">4.3 Linked Style Sheet</a></li>
              <li><a href="4.4 selector.html">4.4 Selector</a></li>
              <li><a href="4.5 Grouping Element.html">4.5 Grouping Element</a></li>
              <li><a href="4.6 Float.html">4.6 Float</a></li>
              <li><a href="4.7 Dropdown Menu.html">4.7 Dropdown Menu</a></li>
            </ul>
          </li>
          <li><a href="AboutMe.html">About Me</a></li>
        </ul>
      </div>
    </div>
  </body>
</html>
```

#### **b. File: 4.7 Dropdown Menu.css**

CSS-nya mengatur posisi menu agar horizontal dan submenu muncul saat diarahkan kursor (hover). Warna dasar menggunakan gradasi ungu dan biru, dengan teks putih agar kontras.

Saat dijalankan, hasilnya:

* Menu utama tampil secara horizontal di tengah halaman.  
* Ketika kursor diarahkan ke **Praktikum 4**, muncul submenu berisi daftar link ke tiap praktikum (4.1–4.7).  
* Warna berubah saat di-hover untuk menunjukkan efek interaktif.  

Berikut isi file CSS-nya:

```css
#wrapper {
  width:600px; 
  margin:auto;
}

#menu ul {
  font-family: Arial, Verdana;
  font-size: 14px;
  margin:0;
  padding:0;
  list-style:none;
}

#menu ul li {
  display: block;
  position: relative;
  float: left;
}

#menu li ul {
  display: none;
}

#menu ul li a {
  display: block;
  text-decoration: none;
  color: #ffffff;
  padding: 5px 15px 5px 15px;
  background: #ff0ff3;
  white-space: nowrap;
}

#menu ul li a:hover {
  background:#6177e6;
}

#menu li:hover ul {
  display: block;
  position: absolute;
}

#menu li:hover li {
  float: none;
  font-size: 11px;
}

#menu li:hover a {
  background:#ff64d1;
}

#menu li:hover li a:hover {
  background: #ff9be1;
}
```
